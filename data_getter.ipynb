```python
import numpy as np
import matplotlib.pyplot as plt
import requests
import json

#https://oda.ft.dk/api/Afstemning?$inlinecount=allpages&$filter=opdateringsdato%20gt%20datetime%272024-11-25T18:52:30.742%27
```

```python
def get_JSON(url):
    
    try:
        # Make a GET request to the API endpoint using requests.get()
        response = requests.get(url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            return response.json()
        else:
            print('Error:', response.status_code)
            return None
    
    except requests.exceptions.RequestException as e:
  
        # Handle any network-related errors or exceptions
        print('Error:', e)
        return None

def paginator(start_url):
    response = get_JSON(start_url)
    entries_per_page = len(response['value'])
    total_entries = int(response['odata.count'])

    values = []
    pages = int(np.ceil(total_entries/entries_per_page))
    for i in range(pages):
        values.append(response['value'])
        if i != (pages-1):
            next_url = response['odata.nextLink']
            response = get_JSON(next_url)
    
    return values
```

Download data for all votes through the API

```python
url = "https://oda.ft.dk/api/Afstemning?$inlinecount=allpages"
response = get_JSON(url)
data_count = int(response['odata.count'])
```

Format it and dump it into a JSON file

```python
values = paginator(url)
with open('ft_votes.json','w') as file:
    json.dump(values,file)
```
